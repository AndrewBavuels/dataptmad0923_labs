For this lab, we are going to be putting together a user similarity based recommender system in a step-by-step fashion. Our data set contains customer grocery purchases, and we will use similar purchase behavior to inform our recommender system. Our recommender system will generate 5 recommendations for each customer based on the purchases they have made.

```python
import pandas as pd
from scipy.spatial.distance import pdist, squareform
```

```python
data = pd.read_csv('../data/customer_product_sales.csv')
```

```python
data.head()
```

## Step 1: Create a data frame that contains the total quantity of each product purchased by each customer.

You will need to group by CustomerID and ProductName and then sum the Quantity field.

```python
total_quantity = data.groupby(['CustomerID','ProductName']).agg({'Quantity':'sum'})
total_quantity.head()
```

## Step 2: Use the `pivot_table` method to create a product by customer matrix.

The rows of the matrix should represent the products, the columns should represent the customers, and the values should be the quantities of each product purchased by each customer. You will also need to replace nulls with zeros, which you can do using the `fillna` method.

```python
product_by_customer = pd.pivot_table(total_quantity, index='ProductName',columns='CustomerID', values='Quantity').fillna(0)
product_by_customer
```

## Step 3: Create a customer similarity matrix using `squareform` and `pdist`. For the distance metric, choose "euclidean."

```python
dist_calculation = pdist(X=product_by_customer.T, metric='euclidean')
len(dist_calculation)
```

```python
# Squareform method: https://scipy.github.io/devdocs/reference/generated/scipy.spatial.distance.squareform.html

dist_distribution = squareform(dist_calculation)
dist_distribution
```

```python
euclid_dist = pd.DataFrame(dist_distribution,
                           index=product_by_customer.columns, 
                           columns=product_by_customer.columns)

euclid_dist
```

## Step 4: Check your results by generating a list of the top 5 most similar customers for a specific CustomerID.

```python
euclid_dist.info()
```

```python
# Top 5 most similar customers for CustomerID == 33

top5_similar_33 = pd.DataFrame(euclid_dist[33].sort_values()[1:6])
# top5_similar_33 = pd.DataFrame(top5_similar_33)
top5_similar_33
```

## Step 5: From the data frame you created in Step 1, select the records for the list of similar CustomerIDs you obtained in Step 4.

```python
# Back to Step 1
type(total_quantity)
```

```python
total_quantity = total_quantity.reset_index()
total_quantity
```

```python
# Back to Step 4
top5_similar_33 = top5_similar_33.reset_index()
top5_similar_33
```

```python
similar_records = top5_similar_33.merge(total_quantity, on='CustomerID').drop(columns=33)
similar_records
```

```python
similar_records['CustomerID'].unique()
```

## Step 6: Aggregate those customer purchase records by ProductName, sum the Quantity field, and then rank them in descending order by quantity.

This will give you the total number of each product purchased by the 5 most similar customers to the customer you selected in order from most purchased to least.

```python
ranking = similar_records.groupby('ProductName').agg({'Quantity': 'sum'}).sort_values('Quantity', ascending=False)
ranking
```

## Step 7: Filter the list for products that the chosen customer has not yet purchased and then recommend the top 5 products with the highest quantities that are left.

- Merge the ranked products data frame with the customer product matrix on the ProductName field.
- Filter for records where the chosen customer has not purchased the product.
- Show the top 5 results.

```python
# Merge the ranked products data frame with the customer product matrix on the ProductName field
recommended_prod = ranking.merge(product_by_customer, on='ProductName').loc[:, ['Quantity',33]]
recommended_prod
```

```python
# Filter for records where the chosen customer has not purchased the product.
recommended_prod = recommended_prod[recommended_prod[33] == 0].sort_values(by='Quantity', ascending= False)
recommended_prod
```

```python
# Show the top 5 results.
recommended_prod = recommended_prod.iloc[0:5]
recommended_prod
```

## Step 8: Now that we have generated product recommendations for a single user, put the pieces together and iterate over a list of all CustomerIDs.

- Create an empty dictionary that will hold the recommendations for all customers.
- Create a list of unique CustomerIDs to iterate over.
- Iterate over the customer list performing steps 4 through 7 for each and appending the results of each iteration to the dictionary you created.

```python
# Create an empty dictionary that will hold the recommendations for all customers
recommend_for_all = {}
```

```python
# Create a list of unique CustomerIDs to iterate over.
customers_list = data['CustomerID'].unique()
type(customers_list)
```

```python
# Iterate over the customer list performing steps 4 through 7 for each and 
# appending the results of each iteration to the dictionary you created.

for customer in customers_list:
    
    # Step 4: Check your results by generating a list of the top 5 most similar customers for a specific CustomerID.
    top5_similar_customer = pd.DataFrame(euclid_dist[customer].sort_values()[1:6])
    top5_similar_customer = top5_similar_customer.reset_index()

    # Step 5: From the data frame you created in Step 1, select the records for the list of similar CustomerIDs you obtained in Step 4.
    similar_records_customer = top5_similar_customer.merge(total_quantity, on='CustomerID').drop(columns=customer)

    # Step 6: Aggregate those customer purchase records by ProductName, sum the Quantity field, and then rank them in descending order by quantity.
    ranking_customer = similar_records_customer.groupby('ProductName').agg({'Quantity': 'sum'}).sort_values('Quantity', ascending=False)

    # Step 7: Filter the list for products that the chosen customer has not yet purchased and then recommend the top 5 products with the highest quantities that are left.
    recommended_products_customer = ranking_customer.merge(product_by_customer, on='ProductName').loc[:, ['Quantity', customer]]
    recommended_products_customer = recommended_products_customer[recommended_products_customer[customer] == 0].sort_values(by='Quantity', ascending=False)
    recommended_products_customer = recommended_products_customer.iloc[0:5]

    # Appending the results of each iteration to the dictionary
    all_recommendations[customer] = recommended_products_customer.index.tolist()
all_recommendations
```

##  Step 9: Store the results in a Pandas data frame. The data frame should a column for Customer ID and then a column for each of the 5 product recommendations for each customer.

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.from_dict.html
top5_recommendations = pd.DataFrame.from_dict(all_recommendations, orient='index').reset_index()
top5_recommendations.columns = ['CustomerID', 'Product_1', 'Product_2', 'Product_3', 'Product_4', 'Product_5']
top5_recommendations.head()
```

```python
top5_recommendations[top5_recommendations['CustomerID'] == 33]
```
