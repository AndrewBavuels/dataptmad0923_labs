### 1. sepal length (cm)

```python
# Setosa

setosa_csv = csv_iris_df[csv_iris_df['class'] == 'setosa']
setosa_sklearn = sklearn_iris_df[sklearn_iris_df['target_name'] == 'setosa']

plt.figure(figsize=(7, 4))

sns.scatterplot(x=setosa_csv.index, y=setosa_csv['sepal length (cm)'], label='CSV', color='blue')
sns.scatterplot(x=setosa_sklearn.index, y=setosa_sklearn['sepal length (cm)'], label='Scikit-learn', color='orange')

plt.title('Scatterplot for Setosas (Sepal Length)')
plt.xlabel('Row Indexes')
plt.ylabel('Sepal Length (cm)')

plt.legend()
plt.tight_layout()

plt.show()
```

The **blue spot** means that record from the csv dataset was modified by increasing the sepal length no more than **5.8 cms**.

```python
# versicolor

versicolor_csv = csv_iris_df[csv_iris_df['class'] == 'versicolor']
versicolor_sklearn = sklearn_iris_df[sklearn_iris_df['target_name'] == 'versicolor']

plt.figure(figsize=(7, 4))

sns.scatterplot(x=versicolor_csv.index, y=versicolor_csv['sepal length (cm)'], label='CSV', color='blue')
sns.scatterplot(x=versicolor_sklearn.index, y=versicolor_sklearn['sepal length (cm)'], label='Scikit-learn', color='orange')

plt.title('Scatterplot for Versicolor (Sepal Length)')
plt.xlabel('Row Indexes')
plt.ylabel('Sepal Length (cm)')

plt.legend()
plt.tight_layout()

plt.show()
```

### Concat for Data Cleaning

```python
csv_iris_df.head()
```

```python
sklearn_iris_df.head()
```

I'm going back up to **review the dataviz with altered records**, to put the correct ones in the csv_iris_df:

- Sepal **length** for setosa, versicolor and virginica
- Sepal **width** for setosa and versicolor (virginica is FINE)
- Petal **length** for setosa (only)
- Petal **width** for setosa and virginica
- Null values for **Petal width** for Iris-setosa

```python
# Sepal length for setosa, versicolor and virginica
sepal_length_df = pd.concat([csv_iris_df['sepal length (cm)'],sklearn_iris_df['sepal length (cm)'],
                            csv_iris_df['class'],sklearn_iris_df['target_name']], axis=1, sort=False)

sepal_length_df['Is True?'] = csv_iris_df['sepal length (cm)'] == sklearn_iris_df['sepal length (cm)']
```

```python
sepal_length_df[sepal_length_df['Is True?'] == False]
```

```python
# Sepal width for setosa and versicolor (virginica is FINE)
sepal_width_df = pd.concat([csv_iris_df['sepal width (cm)'],sklearn_iris_df['sepal width (cm)'],
                            csv_iris_df['class'],sklearn_iris_df['target_name']], axis=1, sort=False)

sepal_width_df['Is True?'] = csv_iris_df['sepal width (cm)'] == sklearn_iris_df['sepal width (cm)']
```

```python
sepal_width_df[sepal_width_df['Is True?'] == False]
```

```python
# Petal length for setosa (only)
petal_length_df = pd.concat([csv_iris_df['petal length (cm)'],sklearn_iris_df['petal length (cm)'],
                            csv_iris_df['class'],sklearn_iris_df['target_name']], axis=1, sort=False)

petal_length_df['Is True?'] = csv_iris_df['petal length (cm)'] == sklearn_iris_df['petal length (cm)']
```

```python
petal_length_df[petal_length_df['Is True?'] == False]
```

```python
# Petal width for setosa and virginica & Null values for Petal width for Iris-setosa

petal_width_df = pd.concat([csv_iris_df['petal width (cm)'],sklearn_iris_df['petal width (cm)'],
                            csv_iris_df['class'],sklearn_iris_df['target_name']], axis=1, sort=False)

petal_width_df['Is True?'] = csv_iris_df['petal width (cm)'] == sklearn_iris_df['petal width (cm)']
```
